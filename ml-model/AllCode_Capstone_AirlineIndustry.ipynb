```python
# Using Combination of Oversampling and Undersampling(SMOTE)

# Checking original class distribution
print("Original class distribution:", Counter(y_train))

# Apply SMOTE
smote = SMOTE(sampling_strategy=0.8, random_state=42)  # 0.8 means minority/majority ratio = 80%
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Check new class distribution
print("Resampled class distribution:", Counter(y_resampled))
```

```python
# Using Oversampling with SMOTE

# Checking original class distribution
print("Original class distribution:", Counter(y_train))

# Applying SMOTE for oversampling
smote = SMOTE(sampling_strategy=0.8, random_state=42)  # Set minority/majority ratio to 0.8
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Checking new class distribution
print("Resampled class distribution (Oversampling):", Counter(y_resampled))
```

```python
#Undersampling with RandomUnderSampler

# Checking original class distribution
print("Original class distribution:", Counter(y_train))

# Applying RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy=0.8, random_state=42)  # Set minority/majority ratio to 0.8
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

# Checking new class distribution
print("Resampled class distribution (Undersampling):", Counter(y_resampled))
```

Random Forest Model

```python
# Random Forest model - top 10

import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_auc_score, classification_report

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Initialize and fit a Random Forest model
rf_model = RandomForestClassifier(random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)  # Fit the model on the training data

# Get feature importance
feature_importances = pd.DataFrame({
    'Features': X.columns,
    'Importance': rf_model.feature_importances_
})

# Sort features by importance and select top 10
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
top_10_features_rf = feature_importances.head(10)['Features'].tolist()

print("Top 10 Features based on Random Forest importance:")
print(top_10_features_rf)

# Subset the training and testing data to include only the top 10 features
X_train_top10 = X_train[top_10_features_rf]
X_test_top10 = X_test[top_10_features_rf]

# Initialize and fit the Random Forest model on the top 10 features
rf_model_top10 = RandomForestClassifier(random_state=42)
rf_model_top10.fit(X_train_top10, y_train)

# Predict on the test set
y_pred_top10 = rf_model_top10.predict(X_test_top10)
y_pred_prob_top10 = rf_model_top10.predict_proba(X_test_top10)[:, 1]

# Calculate Performance Metrics
conf_matrix = confusion_matrix(y_test, y_pred_top10)
tn, fp, fn, tp = conf_matrix.ravel()

sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
gmean = (sensitivity * specificity) ** 0.5
accuracy = (tp + tn) / (tp + tn + fp + fn)
auc = roc_auc_score(y_test, y_pred_prob_top10)

# Print Metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Sensitivity: {sensitivity:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"Precision: {precision:.4f}")
print(f"G-Mean: {gmean:.4f}")
print(f"AUC: {auc:.4f}")

# Print Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_top10))
```
